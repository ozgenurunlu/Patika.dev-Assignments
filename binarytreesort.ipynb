1. Adım: 7'yi kök (root) olarak ekleyelim.
root = 7



```
7
```

2. Adım: 5'i ekleyelim.

5, 7'den küçük olduğu için 7'nin soluna eklenir.



```
    7
   /
  5
```



3. Adım: 1'i ekleyelim.

1, 7'den ve 5'ten de küçük olduğu için 5'in soluna eklenir.



```
    7
   /
  5
 /
1
```



4. Adım: 8'i ekleyelim.

8, 7'den büyük olduğu için 7'nin sağına eklenir.




```
    7
   / \
  5   8
 /
1
```

5. Adım: 3'ü ekleyelim.

3, 7'den ve 5'ten küçük ama 1'den büyük olduğu için 1'in sağına eklenir.


```
    7
   / \
  5   8
 /
1
 \
  3
```

6. Adım: 6'yı ekleyelim.
