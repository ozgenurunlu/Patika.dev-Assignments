# 1. Hangi ülkeler genel mutlulukta en yüksek sırada?

```python

# En mutlu ilk 10 ülke (2015)
top10_2015 = df_2015[['Country', 'Happiness Score']].sort_values(by='Happiness Score', ascending=False).head(10)

# En mutlu ilk 10 ülke (2016)
top10_2016 = df_2016[['Country', 'Happiness Score']].sort_values(by='Happiness Score', ascending=False).head(10)

# En mutlu ilk 10 ülke (2017)
top10_2017 = df_2017[['Country', 'Happiness.Score']].sort_values(by='Happiness.Score', ascending=False).head(10)

print("2015 En Mutlu Ülkeler:\n", top10_2015)
print("2016 En Mutlu Ülkeler:\n", top10_2016)
print("2017 En Mutlu Ülkeler:\n", top10_2017)
```

# 2. Mutluluğa katkıda bulunan 6 faktör bakımından en yüksek skora sahip ülkeler

```python
# 2017 için faktörlerde en yüksek skora sahip ülkeler
columns_2017 = {
    'Ekonomik üretim': 'Economy..GDP.per.Capita.',
    'Sosyal destek': 'Family',
    'Sağlıklı yaşam': 'Health..Life.Expectancy.',
    'Özgürlük': 'Freedom',
    'Yolsuzluk': 'Trust..Government.Corruption.',
    'Cömertlik': 'Generosity'
}

for faktör, kolon in columns_2017.items():
    en_yuksek = df_2017[['Country', kolon]].sort_values(by=kolon, ascending=False).head(1)
    print(f"🔹 {faktör} açısından en yüksek ülke:\n{en_yuksek}\n")
```

# 3. Hangi bölge/continent ortalama olarak daha mutlu?

```python
df_2015 = pd.read_csv('/kaggle/input/world-happiness/2015.csv')
df_2015.columns = df_2015.columns.str.replace('[^A-Za-z0-9]+', '_', regex=True)

region_mean = df_2015.groupby('Region')['Happiness_Score'].mean().sort_values(ascending=False)
print("2015 Bölgelere Göre Ortalama Mutluluk Puanları:\n", region_mean)

# Görselleştirme
plt.figure(figsize=(12,6))
sns.barplot(x=region_mean.values, y=region_mean.index, palette="viridis")
plt.title('2015 Bölgelere Göre Ortalama Mutluluk Skoru')
plt.xlabel('Ortalama Mutluluk Skoru')
plt.ylabel('Bölge')
plt.show()
```

# 4.Cömertlik ile mutluluk arasında nasıl bir ilişki var?

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_2015, x='Generosity', y='Happiness_Score')
plt.title('2017: Cömertlik ile Mutluluk Skoru Arasındaki İlişki')
plt.xlabel('Cömertlik')
plt.ylabel('Mutluluk Skoru')
plt.show()

corr = df_2015['Generosity'].corr(df_2015['Happiness_Score'])
print(f"Cömertlik ve Mutluluk Korelasyonu: {corr:.3f}.")
```

Grafikte gördüğümüz üzere, ülkelerin cömertlik skorları ile mutluluk skorları arasında pozitif bir ilişki bulunmakla birlikte bu ilişki çok kuvvetli değil. Korelasyon katsayısı genellikle 0.2–0.3 civarında çıkar, yani cömertlik mutluluğu artıran faktörlerden biri ancak tek başına mutluluğu güçlü şekilde açıklamıyor.

Bu durum, insanların daha mutlu olmalarında ekonomik faktörler, sosyal destek ve sağlık gibi başka unsurların da önemli olduğunu gösteriyor. Cömertliğin etkisi, toplumsal dayanışma ve sosyal ilişkilerin kalitesini artırarak mutluluğa katkı sağlıyor olabilir.

# 5.En cömert ülkeler ile en mutlu ülkeler örtüşüyor mu?

```python
top_generous = set(df_2015.sort_values(by='Generosity', ascending=False).head(10)['Country'])
top_happy = set(df_2015.sort_values(by='Happiness_Score', ascending=False).head(10)['Country'])

print("En cömert ülkeler:", top_generous)
print("En mutlu ülkeler:", top_happy)
print("İki kümenin kesişimi (hem en cömert hem en mutlu):", top_generous.intersection(top_happy))
```

# 6.Yolsuzluğun en az olduğu ülkeler mutlulukta üst sıralarda mı?

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_2017, x='Trust..Government.Corruption.', y='Happiness.Score')
plt.title('2017: Yolsuzluğun Azlığı ve Mutluluk Skoru')
plt.xlabel('Yolsuzluğun Azlığı (Trust)')
plt.ylabel('Mutluluk Skoru')
plt.show()

corr_trust = df_2017['Trust..Government.Corruption.'].corr(df_2017['Happiness.Score'])
print(f"Yolsuzluğun Azlığı ve Mutluluk Korelasyonu: {corr_trust:.3f}")
```

2017 verisine göre, yolsuzluğun daha az olduğu ülkelerde mutluluk skorlarının genel olarak daha yüksek olduğu gözlemlenmektedir. Ancak bu ilişki güçlü ve doğrudan bir korelasyon göstermemektedir.

Grafikteki noktalar dağınık bir yapıya sahiptir, bu da “Trust (Government Corruption)” değişkeninin mutluluğu etkileyen faktörlerden biri olduğunu, ancak tek başına yeterli olmadığını gösterir.
