# 3. Hangi bÃ¶lge/continent ortalama olarak daha mutlu?

```python
df_2015 = pd.read_csv('/kaggle/input/world-happiness/2015.csv')
df_2015.columns = df_2015.columns.str.replace('[^A-Za-z0-9]+', '_', regex=True)

region_mean = df_2015.groupby('Region')['Happiness_Score'].mean().sort_values(ascending=False)
print("2015 BÃ¶lgelere GÃ¶re Ortalama Mutluluk PuanlarÄ±:\n", region_mean)

# GÃ¶rselleÅŸtirme
plt.figure(figsize=(12,6))
sns.barplot(x=region_mean.values, y=region_mean.index, palette="viridis")
plt.title('2015 BÃ¶lgelere GÃ¶re Ortalama Mutluluk Skoru')
plt.xlabel('Ortalama Mutluluk Skoru')
plt.ylabel('BÃ¶lge')
plt.show()
```

# 4.CÃ¶mertlik ile mutluluk arasÄ±nda nasÄ±l bir iliÅŸki var?

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_2015, x='Generosity', y='Happiness_Score')
plt.title('2017: CÃ¶mertlik ile Mutluluk Skoru ArasÄ±ndaki Ä°liÅŸki')
plt.xlabel('CÃ¶mertlik')
plt.ylabel('Mutluluk Skoru')
plt.show()

corr = df_2015['Generosity'].corr(df_2015['Happiness_Score'])
print(f"CÃ¶mertlik ve Mutluluk Korelasyonu: {corr:.3f}.")
```

Grafikte gÃ¶rdÃ¼ÄŸÃ¼mÃ¼z Ã¼zere, Ã¼lkelerin cÃ¶mertlik skorlarÄ± ile mutluluk skorlarÄ± arasÄ±nda pozitif bir iliÅŸki bulunmakla birlikte bu iliÅŸki Ã§ok kuvvetli deÄŸil. Korelasyon katsayÄ±sÄ± genellikle 0.2â€“0.3 civarÄ±nda Ã§Ä±kar, yani cÃ¶mertlik mutluluÄŸu artÄ±ran faktÃ¶rlerden biri ancak tek baÅŸÄ±na mutluluÄŸu gÃ¼Ã§lÃ¼ ÅŸekilde aÃ§Ä±klamÄ±yor.

# 5.En cÃ¶mert Ã¼lkeler ile en mutlu Ã¼lkeler Ã¶rtÃ¼ÅŸÃ¼yor mu?

```python
top_generous = set(df_2015.sort_values(by='Generosity', ascending=False).head(10)['Country'])
top_happy = set(df_2015.sort_values(by='Happiness_Score', ascending=False).head(10)['Country'])

print("En cÃ¶mert Ã¼lkeler:", top_generous)
print("En mutlu Ã¼lkeler:", top_happy)
print("Ä°ki kÃ¼menin kesiÅŸimi (hem en cÃ¶mert hem en mutlu):", top_generous.intersection(top_happy))
```
